```python
import tensorflow as tf
```

```python
vector=tf.constant([1,2,3,4,5,6])
```

```python
print('shape:',vector.shape)
```

```python
print('dimension:',vector.ndim)
```

```python
matrix=tf.reshape(vector,shape=(2,3))
print(matrix.numpy())
```

```python
print('original shape:',vector.shape)
print('New Shape:',matrix.shape)
```

```python
print('Original Dimension:',vector.ndim)
print('New Dimension:',matrix.ndim)
```

```python
new_matrix=tf.reshape(vector,shape=(3,2))
print(new_matrix.numpy())
```
