```python
import tensorflow as tf
```

```python
print('version:',tf.__version__)
```

```python
scaler=tf.constant(5)
print(scaler)
```

```python
print("number only:",scaler.numpy())
```

```python
print('shape:',scaler.shape)
```

```python
print('dimension:',scaler.ndim)
```

```python
vector=tf.constant([1,2,3])
print('vector:',vector.numpy())
```

```python
print('shape:',vector.shape)
```

```python
print('dimension:',vector.ndim)
```

```python
matrix=tf.constant([[1,2,3],
[4,5,6]])
print(matrix)
```

```python
print('values:',matrix.numpy())
```

```python
print('shape:',matrix.shape)
```

```python
print('dimension:',matrix.ndim)
```

```python
print('first column:',matrix[:,0].numpy())
```

```python
tensor_3d=tf.constant([[[1,2,3],
[4,5,6]],

[[5,5,6],
[9,0,1]]])
```

```python
print('shape:',tensor_3d.shape)
```

```python
print('Dimensions:',tensor_3d)
```

```python
print('values',tensor_3d.numpy())
```

```python
print('second matrix second column:',tensor_3d[1][:,2])
```

```python
print('values:',tensor_3d[1][:,2].numpy())
```

```python
print(tensor_3d[1,0].numpy())
```
